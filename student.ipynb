# Data Cleaning

```python
#Identify missing values across dataframe columns 

df.isna().sum()/len(df)*100
```

```python
# Copy of my Original data
original_data = df.copy()
```

```python
df = original_data.copy()
```

```python
#Identify columns with more than 50% of missing data and add them to a list

columns_to_drop=[]
for x in df:
    if (df[x].isna().sum() / len(df[x])) * 100 > 50:
        columns_to_drop.append(x)
print(columns_to_drop)
```

```python
#Drop columns from dataframe that are missing too much data
df= df.drop(columns= columns_to_drop)
df
```

```python
# idenitify irrelevant columns and drop them in this case we will only reamin with columns which will be useful for data analysis
columns =['Investigation.Type','Accident.Number', 'Airport.Code', 'Airport.Name','Registration.Number']
df=df.drop(columns,axis=1)
df
```

```python
#Calculating the mode of the columns having numerical datatypes.
modes = df[['Number.of.Engines', 'Total.Fatal.Injuries', 'Total.Serious.Injuries', 'Total.Minor.Injuries', 'Total.Uninjured']].mode()
modes
```

```python
#fill the numerical values with mode
for column in ['Number.of.Engines', 'Total.Fatal.Injuries', 'Total.Serious.Injuries', 'Total.Minor.Injuries', 'Total.Uninjured']:
    df[column].fillna(df[column].mode()[0], inplace=True)
```

```python
#fill the missing values in the reamining columns using forward and backward fill
df.fillna(method='ffill').fillna(method='bfill')
```

```python
df.isna().sum()/len(df)*100
```

```python
# Converting 'Event.Date' and 'Publication.Date' to datetime datatype
df['Event.Date'] = pd.to_datetime(df['Event.Date'], errors='coerce')
df['Publication.Date'] = pd.to_datetime(df['Publication.Date'], errors='coerce', format='%d-%m-%Y')
```

# Exploratory Data Analysis /Data Visualization

# Count the number of Accidents as per Countries


```python
# Count of Accidents as per Countries
accidents =df['Country'].groupby(df['Country']).count()
accidents
```

# Plot 1: Distribution of Accidents in different Countries (Count of Accidents > 200)

```python
fig = plt.figure(figsize=(15,20))
accidents_Count = accidents[df['Country'].groupby(df['Country']).count() > 200]
accidents_Count_df = pd.DataFrame({'Country':accidents_Count.index,'Count':accidents_Count.values})
plt.bar(accidents_Count_df['Country'], height = accidents_Count_df['Count'], color='red')
plt.xticks(rotation=90)
plt.xlabel("Country", size=15)
plt.ylabel("Count of Accidents", size=15)
y=accidents_Count_df['Count']
for i,v in enumerate(y):
    plt.text(x=i, y=v, s=str(v), horizontalalignment='center', size=15)
plt.title("Distribution of Accidents wrt Countries", size=20)
plt.show()
```

### from the above plot, there are 7 countries namely Australia, Bahamas, Brazil, Canada, Mexico, United Kingdom and the United States which have count of accidents more than 200. Also, most of the accidents have occurred in the US only with staggering 82,248count

# Plot 2: Top 20 Companies which are involved in accidents

```python
# Evaluate the Companies which are common in accidents

plt.figure(figsize=(15,10))
df['Make'].str.upper().value_counts().sort_values(ascending=False)[:20].plot(kind='bar', color='Green')
plt.xticks(rotation=90)
plt.xlabel("Make", size=15)
plt.ylabel("Count of Accidents", size=15)
plt.title("Top 20 Aircraft Companies in the United States", size=20)
y=df['Make'].str.upper().value_counts().sort_values(ascending=False)[:20]
for i, v in enumerate(y):
    plt.text(i, v, str(v), fontsize=15, style='oblique', horizontalalignment='center')
plt.show()
```

### In this plot, we are analysing the top 20  companies which are involved in most of the accidents. 'Cessna' has 25,853 accidents in the bucket whereas 'Piper' and ' Beech' have 14,168 and 2285 accidents, respectively.

# Plot 3: What was the Phase of Flight and Weather Conditions when accident took place?

```python
plt.figure(figsize=(15,10))
splot = sns.countplot(data=df, x="Broad.phase.of.flight", hue='Weather.Condition', edgecolor=(0,0,0), linewidth=1, palette='Reds')
plt.xticks(rotation=90)
plt.xlabel("Broad Phase of Flight", size=15)
plt.ylabel("Number of Occurrences", size=15)
plt.title("What was the Phase of Flight when Accident took place?", size=20)
for p in splot.patches:
    splot.annotate(format(p.get_height()), (p.get_x() + p.get_width() / 2., 
    p.get_height()), ha = 'center', va = 'bottom', xytext = (0, 10), textcoords = 'offset points', rotation=90, size=11.5)
plt.show()
```

### This plot illustrates that most of the accidents took place when aircraft was either Landing or Taking off. 14,902 accidents took place when the airplace was Landing and the weather conditions were clear for flight. About 15,385 accidents took place during take off with clear weather conditions.

# Plot 4: What was the Engine Type of Fatal or Non-Fatal accidents?

```python
cplot = sns.catplot(x="Engine.Type", col="Injury.Severity", kind="count", col_wrap=2,
            data=df[df['Injury.Severity'].isin(['Fatal','Non-Fatal'])], height=4,
                 legend_out=True, size=8, palette="Set3", edgecolor=(0,0,0))
cplot.set_xticklabels(rotation=90)
cplot.fig.suptitle("What was the Engine Type of Fatal or Non-Fatal accidents?", size=20)
cplot.fig.subplots_adjust(top=.9)
cplot.set_titles("Injury Severity Type is {col_name}")
cplot.set(ylim=[0,65000])
for ax in cplot.axes.ravel():
    for p in ax.patches:
        ax.annotate(format(p.get_height()), (p.get_x() + p.get_width() / 2, 
        p.get_height()), ha = 'center', va = 'bottom', xytext = (0, 10), textcoords = 'offset points', rotation=90, size=12)
plt.show()
```

##### Post analysis of above two plots, it can said that most of the Fatal or Non-Fatal accident occurred for Engine Type 'Reciprocating'. For Fatal, count is 13,068 whereas for Non-Fatal, count is around 54,310.
